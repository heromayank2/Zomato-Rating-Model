```python
# Importing the Libraries

import pandas as pd
import numpy as np
```

```python
# Importing Dataset
train = pd.read_csv('train.csv', encoding = "ISO-8859-1")
test =pd.read_csv('test.csv', encoding = "ISO-8859-1")
y=train.iloc[:,15]
```

```python
# Dropping Off Unused ColumnS

x=train.drop(['Restaurant Name','Locality','Country Code','Restaurant ID','Address','Locality Verbose', 'Aggregate rating'], axis=1)
test=test.drop(['Restaurant Name','Locality','Country Code','Restaurant ID','Address','Locality Verbose'],axis=1)
```

```python
# Concatinating train and test data
frames = [x, test]
x = pd.concat(frames)
```

```python
x.head()
```

```python
# Creating New Columns(features) X, Y, Z Using Longitude and Latitude Given

Longitude = x['Longitude']
Longitude=pd.to_numeric(Longitude)
mean_longitude = Longitude.mean(skipna=True)
Longitude=Longitude.replace(0,mean_longitude) 
Longitude_COS=np.cos(Longitude)
Longitude_SIN=np.sin(Longitude)

Latitude = x['Latitude']
Latitude=pd.to_numeric(Latitude)
mean_latitude = Latitude.mean(skipna=True)
Latitude=Latitude.replace(0,mean_latitude)
Latitude_COS=np.cos(Latitude)
Latitude_SIN=np.sin(Latitude)
```

```python
CordinateX=Longitude_COS*Latitude_COS
CordinateY=Longitude_SIN*Latitude_COS
CordinateZ=Latitude_SIN
```

```python
# Adding Columns to Total Dataset

x['CordinateX']=CordinateX
x['CordinateY']=CordinateY
x['CordinateZ']=CordinateZ
```

```python
x.head()
```

```python
# Dropping Off Unused Columns

x=x.drop(['Latitude','Longitude'],axis=1)
```

```python
x.head()
```

```python
# Label Encoding Of Columns

from sklearn import preprocessing
TB = preprocessing.LabelEncoder()
OD = preprocessing.LabelEncoder()
CR = preprocessing.LabelEncoder()
TB.fit(x['Has Table booking'])
OD.fit(x['Has Online delivery'])
CR.fit(x['Currency'])
```

```python
#Transforming obvious Columns

x['Has Table booking']=TB.transform(x['Has Table booking'])
x['Has Online delivery']=OD.transform(x['Has Online delivery'])
x['Currency']=CR.transform(x['Currency'])
```

```python
x.head()
```

```python
#Extracting Cuisines from Dataset

foodlist=[]
for food in x['Cuisines']:
    fooditems=food.split(", ")
    print(fooditems)
    foodlist.extend(fooditems)
```

```python
foodlist=list(set(foodlist))
```

```python
FOODPORN=dict((food,0) for food in foodlist)
```

```python
for food in x['Cuisines']:
    fooditems=food.split(", ")
    print(fooditems)
    for item in fooditems:
        FOODPORN[item]+=1
```

```python
#Creating New Column as Food Score

foodscore=[]
i=0
for food in x['Cuisines']:
    score=0
    fooditems=food.split(", ")
    print(fooditems)
    i+=1
    for item in fooditems:
        score+=FOODPORN[item]
    foodscore.append(score)    
```

```python
len(foodscore)
```

```python
x['Foodscore']=foodscore
```

```python
x.head()
```

```python
city_list=x['City'].value_counts()
```

```python
city_list['New Delhi']
```

```python
#Creating New Column as City Popularity 

city_score=[]
for city in x['City']:
    city_score.append(city_list[city])
```

```python
city_category=[]
for city in city_score:
    if(10>=city>0):
        city_category.append(1)
    elif(100>city>10):
        city_category.append(2)
    elif(500>city>=100):
        city_category.append(3)
    elif(2000>city>=500):
        city_category.append(4)    
    elif(city>2000):
        city_category.append(5)    
```

```python
x['City Popularity']=city_category
```

```python
x=x.drop(['Cuisines','City'],axis=1)
```

```python
TrainData=x[:9471]
TestData=x[9471:]
```

```python
TrainData.head()
```

```python
y=np.array(list(y)).reshape((len(y),))
```

```python
#Training Model

from sklearn.ensemble import RandomForestRegressor
clf = RandomForestRegressor(n_estimators=250,random_state=42)
clf.fit(TrainData, y)
```

```python
results=clf.predict(TestData)
```
